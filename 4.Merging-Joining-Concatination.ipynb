```python
import numpy as np
import pandas as pd
```

**Merging 2 Dataframes**

```python
employee= {
    'employee_id':  [1,2,3,4,5],
    'name': ['sourabha','harshith','priya','preethi','bob'],
    'department': ['CEO','HR','IT','finance','HR']
}
employee=pd.DataFrame(employee)
```

```python
salaries= {
    'employee_id': [1,2,3,6,7],
    'salary': [110000,100000,50000,60000,40000],
    'bonus':[15000,10000,1000,1200,1300]
}
salary=pd.DataFrame(salaries)
```

```python
employee
```

```python
salary
```

```python
pd.merge(employee,salary)
```

```python
pd.merge(employee,salary, on='employee_id')
```

```python
pd.merge(employee,salary, on='employee_id',how='inner')
```

```python
pd.merge(employee,salary, on='employee_id',how='outer')
```

```python
pd.merge(employee,salary, on='employee_id',how='left')
```

```python
pd.merge(employee,salary, on='employee_id',how='right')
```

**Concatination of 2 Dataframes**

```python
df1 = pd.DataFrame({
    'A': ['A0','A1','A2'],
    'B': ['B0','B1','B2'],
    'C': ['C0','C1','C2']
})
```

```python
df2 = pd.DataFrame({
    'A': ['A3','A4','A5'],
    'B': ['B3','B4','B5'],
    'C': ['C3','C4','C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df2,df1])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],axis=0)
```

**Joining 2 Dataframes**

```python
df1 = pd.DataFrame({
    'name': ['alice','bob','charlie']
}, index=[1,2,3])
```

```python
df2 = pd.DataFrame({
    'score': [85,90,75]
}, index=[2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2,how='outer')
```

```python
df2.join(df1)
```
