```python
import numpy as np
import pandas as pd
```

**CREATING A DATAFRAME**

```python
#using dict
data_dict = {                
    'name': ['a','b','c'],
    'age': [18,19,20],
    'city': ['bangalore','tumkur','mangalore'],
    'salary': [70000,100000,50000]
}
df = pd.DataFrame(data_dict)
df
```

```python
#using list
data_list= [
    ['a',18,'bangalore',70000],
    ['b',19,'tumkur',100000],
    ['c',20,'mangalore',50000]
]
df2= pd.DataFrame(data_list)
col= ["name","age","city","salary"]
df2= pd.DataFrame(data_list,columns=col)
df2
```
