**PANDAS SERIES**

**Creating series**

```python
import numpy as np
import pandas as pd
```

```python
labels= ['a','b','c']
my_list= [10,20,30]
arr= np.array([10,20,30])
d= {'a':10,'b':20,'c':30}
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list,index=labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(arr,index=labels)
```

```python
pd.Series(d)
```
